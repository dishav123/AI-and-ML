Task - 1:

Create a Python program that converts between different units of measurement.

```python
def convert_units():
    """
    Convert units for length, weight or volume
    """

    print("Enter conversion type:")
    print("1 = Length")
    print("2 = Weight")
    print("3 = Volume")

    choice = input("Enter your choice: ")

    try:
        value = float(input("Enter a value to convert: "))

        if choice == "1":
            print("1 = meters to feet")
            print("2 = feet to meters")
            option = input("Choose an option: ")

            if option == "1":
                result = value * 3.28084
                print("Result:", result, "feet")

            elif option == "2":
                result = value / 3.28084
                print("Result:", result, "meters")

        elif choice == "2":
            print("1 = kg to pounds")
            print("2 = pounds to kg")
            option = input("Choose an option: ")

            if option == "1":
                result = value * 2.20462
                print("Result:", result, "pounds")

            elif option == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == "3":
            print("1 = liters to gallons")
            print("2 = gallons to liters")
            option = input("Choose an option: ")

            if option == "1":
                result = value * 0.264172
                print("Result:", result, "gallons")

            elif option == "2":
                result = value / 0.264172
                print("Result:", result, "liters")

        else:
            print("Wrong choice")

    except:
        print("Please enter a valid number")

convert_units()
```

Task - 2:

Create a Python program that performs various mathematical operations on a list of numbers.

```python
def find_sum(numbers):
    """Return sum of numbers"""
    total = 0
    for n in numbers:
        total = total + n
    return total


def find_avg(numbers):
    """Return average"""
    total = find_sum(numbers)
    avg = total / len(numbers)
    return avg


def find_max(numbers):
    """Return maximum value"""
    maximum = numbers[0]
    for n in numbers:
        if n > maximum:
            maximum = n
    return maximum


def find_min(numbers):
    """Return minimum value"""
    minimum = numbers[0]
    for n in numbers:
        if n < minimum:
            minimum = n
    return minimum


try:
  user_input = input("Enter numbers separated by space: ")
  numbers = []

  for n in user_input.split():
    numbers.append(float(n))

  print("Choose the operation to perform:")
  print("1 = Sum")
  print("2 = Average")
  print("3 = Maximum")
  print("4 = Minimum")

  choice  = int(input("Enter your choice: "))

  if choice == 1:
    print(find_sum(numbers))

  elif choice == 2:
    print(find_avg(numbers))

  elif choice == 3:
    print(find_max(numbers))

  elif choice == 4:
    print(find_min(numbers))

  else:
    print("Invalid choice")

except:
    print("Invalid input")
```

# **4.2 Exercise on List Manipulation:**

1. Extract Every Other Element:

Write a Python function that extracts every other element from a list, starting from the first element.

```python
def extract_every_other(lst):
    result = []
    for i in range(0, len(lst), 2):
        result.append(lst[i])
    return result
```

2. Slice a Sublist:

Write a Python function that returns a sublist from a given list, starting from a specified index and
ending at another specified index.

```python
def get_sublist(lst, start, end):
    return lst[start:end+1]
```

3. Reverse a List Using Slicing:

Write a Python function that reverses a list using slicing

```python
def reverse_list(lst):
    return lst[::-1]
```

4. Remove the First and Last Elements:

Write a Python function that removes the first and last elements of a list and returns the resulting
sublist.

```python
def remove_first_last(lst):
    return lst[1:-1]
```

5. Get the First n Elements:

Write a Python function that extracts the first n elements from a list.

```python
def get_first_n(lst, n):
    return lst[:n]
```

6. Extract Elements from the End:

Write a Python function that extracts the last n elements of a list using slicing.

```python
def get_last_n(lst, n):
    return lst[-n:]
```

7. Extract Elements in Reverse Order:

Write a Python function that extracts a list of elements in reverse order starting from the second-to-last
element and skipping one element in between.

```python
def reverse_skip(lst):
    return lst[-2::-2]
```

**Test Cases**

```python
test_list = [41, 52, 36, 14, 92, 18]

print("Extract Every Other Element:", extract_every_other(test_list))
print("Slice a Sublist:", get_sublist(test_list, 1, 3))
print("Reverse a List Using Slicing:", reverse_list(test_list))
print("Remove the First and Last Elements:", remove_first_last(test_list))
print("Get the First n Elements:", get_first_n(test_list, 3))
print("Extract Elements from the End:", get_last_n(test_list, 3))
print("Extract Elements in Reverse Order:", reverse_skip(test_list))
```

# **4.3 Exercise on Nested List:**

1. Flatten a Nested List:

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

```python
def flatten(lst):
    result = []
    for sublist in lst:
        for item in sublist:
            result.append(item)
    return result
```

2. Accessing Nested List Elements:

Write a Python function that extracts a specific element from a nested list given its indices.

```python
def access_nested_element(lst, indices):
    element = lst

    for i in indices:
        element = element[i]

    return element
```

3. Sum of All Elements in a Nested List:

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).


```python
def sum_nested(lst):
    total = 0

    for item in lst:
        if type(item) == list:
            total = total + sum_nested(item)
        else:
            total = total + item

    return total
```

4. Remove Specific Element from a Nested List:

Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(lst, elem):
    result = []

    for item in lst:
        if type(item) == list:
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)

    return result
```

5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    max_val = None

    for item in lst:
        if type(item) == list:
            value = find_max(item)
        else:
            value = item

        if max_val is None or value > max_val:
            max_val = value

    return max_val
```

6. Count Occurrences of an Element in a Nested List:

Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    count = 0

    for item in lst:
        if type(item) == list:
            count = count + count_occurrences(item, elem)
        elif item == elem:
            count = count + 1

    return count
```

7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    result = []

    for item in lst:
        if type(item) == list:
            result.extend(deep_flatten(item))
        else:
            result.append(item)

    return result
```

7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)
```

**Test Cases**

```python
nested_list = [[1,2],[3,[4,5]],6]

print("Flatten:", flatten([[1,2],[3,4],[5]]))
print("Access Nested Element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1,2]))
print("Sum Nested:", sum_nested(nested_list))
print("Remove Element:", remove_element([[1,2],[3,2],[4,5]], 2))
print("Find Max:", find_max(nested_list))
print("Count Occurrences:", count_occurrences([[1,2],[2,3],[2,4]], 2))
print("Deep Flatten:", deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
print("Average Nested:", average_nested([[1,2],[3,4],[5,6]]))
```

Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9

```python
import numpy as np

A = np.array([
    [2,-3,1],
    [1,-1,2],
    [3,1,-1]
])

B = np.array([-1,-3,9])

A_inv = np.linalg.inv(A)

X = np.dot(A_inv, B)

print("Solution using inverse method:", X)
```

# **10.2 Experiment: How Fast is Numpy?**

**1. Element-wise Addition:**

```python
import time

size = 1000000

# python lists
a = list(range(size))
b = list(range(size))

start = time.time()

result = []
for i in range(size):
    result.append(a[i] + b[i])

end = time.time()

print("Python addition time:", end-start)
```

```python
# NumPy version

a = np.arange(size)
b = np.arange(size)

start = time.time()

result = a + b

end = time.time()

print("NumPy addition time:", end-start)
```

**2. Element-wise Multiplication:**

```python
import time

size = 1000000

# python lists
a = list(range(size))
b = list(range(size))

start = time.time()

result = []
for i in range(size):
    result.append(a[i] * b[i])

end = time.time()

print("Python multiplication time:", end-start)
```

```python
# NumPy version

a = np.arange(size)
b = np.arange(size)

start = time.time()

result = a * b

end = time.time()

print("NumPy multiplication time:", end-start)
```

**3. Dot Product**

```python
import time

size = 1000000

a = list(range(size))
b = list(range(size))

start = time.time()

dot = 0
for i in range(size):
    dot = dot + a[i] * b[i]

end = time.time()

print("Python dot product:", dot)
print("Python time:", end - start)
```

**4. Matrix Multiplication**

```python
import time

A = [[1,2],[3,4]]
B = [[5,6],[7,8]]

start = time.time()

result = [[0,0],[0,0]]

for i in range(2):
    for j in range(2):
        for k in range(2):
            result[i][j] += A[i][k] * B[k][j]

end = time.time()

print("Python matrix multiplication:", result)
print("Python time:", end - start)
```
